```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.font_manager import FontProperties
import matplotlib.dates as mdates    #處理日期

myfont=FontProperties(fname=r'./ttf/NotoSansCJKtc-Black.otf',size=14)    #繁體
sns.set(font=myfont.get_family())
sns.set_style("whitegrid",{"font.sans-serif":['Microsoft JhengHei']})

plt.figure(figsize=(24,8))
plt.title("時間作圖",fontsize=24)
x = ['2021-03-14 01:00:00','2021-03-14 02:00:00','2021-03-14 03:00:00','2021-03-14 04:00:00','2021-03-14 05:00:00',
    '2021-03-14 06:00:00','2021-03-14 07:00:00','2021-03-14 08:00:00','2021-03-14 09:00:00','2021-03-14 10:00:00',
    '2021-03-14 11:00:00','2021-03-14 12:00:00','2021-03-14 13:00:00','2021-03-14 14:00:00','2021-03-14 15:00:00',
    '2021-03-14 16:00:00','2021-03-14 17:00:00','2021-03-14 18:00:00','2021-03-14 19:00:00','2021-03-14 20:00:00']
x = pd.to_datetime(x)    #must
y = ['32','33','34','35','36','32','33','34','35','37',
    '30','34','32','31','30','31','33','31','32','34',]

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M:%S'))  #設定x軸主刻度顯示格式（日期）
plt.gca().xaxis.set_major_locator(mdates.HourLocator(interval=5))  #設定x軸主刻度間距，每小時
plt.plot(x,y,color='blue',label='values')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Time',fontsize=24)
plt.scatter(x,y, color='black')
plt.legend(bbox_to_anchor=(0.9, 0.7), loc=2,ncol= 1, prop={'size': 16})
# plt.savefig('./images/Temp_'+str(i)+'.jpg',dpi=300)   #儲存圖片
```
