```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('./images/lena.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)    #轉灰階
blurred = cv2.GaussianBlur(gray, (5, 5), 0)    #影像去雜訊
canny = cv2.Canny(blurred, 40, 150)    #邊緣偵測

# threshold指令讓我們執行Thresholding處理，它需要四個參數：第一個是待處理的相片。
# 第二個是threshold值（我們暫稱它為T值）。第三個是最大值，亦即當某點超過T值時則設為該最大值，
# 由於我們想要全黑白相片，故此值設為255。第四個為Thresholding的處理方式，表示計算後輸出後影像的呈現型態，
# 針對此參數openCV提供THRESH_BINARY、THRESH_BINARY_INV、THRESH_TRUNC、THRESH_TOZERO、THRESH_TOZERO_INV等五種
(T, thresh) = cv2.threshold(blurred, 95, 255, cv2.THRESH_BINARY)

fig=plt.figure(figsize=(32, 32))
fig.add_subplot(1,5,1)
plt.imshow(image)

fig=plt.figure(figsize=(32, 32))
fig.add_subplot(1,5,2)
plt.imshow(gray)

fig=plt.figure(figsize=(32, 32))
fig.add_subplot(1,5,3)
plt.imshow(blurred)

fig=plt.figure(figsize=(32, 32))
fig.add_subplot(1,5,4)
plt.imshow(canny)

fig=plt.figure(figsize=(32, 32))
fig.add_subplot(1,5,5)
plt.imshow(thresh)

plt.show()
```
