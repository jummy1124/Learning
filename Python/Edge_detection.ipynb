```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('./images/lena.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)    #
blurred = cv2.GaussianBlur(gray, (5, 5), 0)    #
canny = cv2.Canny(blurred, 40, 150)    #

# thresholdThresholding
# thresholdTT
# 255Thresholding
# openCVTHRESH_BINARYTHRESH_BINARY_INVTHRESH_TRUNCTHRESH_TOZEROTHRESH_TOZERO_INV
(T, thresh) = cv2.threshold(blurred, 95, 255, cv2.THRESH_BINARY)

fig=plt.figure(figsize=(32, 32))
fig.add_subplot(1,5,1)
plt.imshow(image)

fig=plt.figure(figsize=(32, 32))
fig.add_subplot(1,5,2)
plt.imshow(gray)

fig=plt.figure(figsize=(32, 32))
fig.add_subplot(1,5,3)
plt.imshow(blurred)

fig=plt.figure(figsize=(32, 32))
fig.add_subplot(1,5,4)
plt.imshow(canny)

fig=plt.figure(figsize=(32, 32))
fig.add_subplot(1,5,5)
plt.imshow(thresh)

plt.show()
```
