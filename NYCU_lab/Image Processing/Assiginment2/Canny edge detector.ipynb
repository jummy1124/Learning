```python
import cv2 
import numpy as np
import canny_edge_detector as ced
from matplotlib import pyplot as plt
```

<h2>Gray images</h2>

<hr>
<h3>Experiment one : coins</h3>

```python
img = cv2.imread('./pic4PR2/coins.jpg',0)
print("Image size:",img.shape)
blurred = cv2.GaussianBlur(img, (5, 5), 0)
edges = cv2.Canny(blurred,30,150)
detector = ced.cannyEdgeDetector(img, sigma=1.4, kernel_size=5, lowthreshold=0.09, highthreshold=0.17, weak_pixel=100)
imgs_final = detector.detect(1)
plt.figure(figsize=(16,8))
plt.subplot(131),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(edges,cmap = 'gray')
plt.title('Canny Edge Image(opencv result)'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(imgs_final[0],cmap = 'gray')
plt.title('Canny Edge Image(my implement)'), plt.xticks([]), plt.yticks([])
plt.show()
```

<h3>Experiment five : rose</h3>

```python
img = cv2.imread('./pic4PR2/rose.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print("Image size:",img_rgb.shape)
edges = cv2.Canny(img_rgb,0,35)
detector = ced.cannyEdgeDetector(img_rgb, sigma=1.4, kernel_size=5, lowthreshold=0.001, highthreshold=0.03, weak_pixel=100)
imgs_final = detector.detect(img_rgb.shape[2])
plt.figure(figsize=(24,12))
plt.subplot(231),plt.imshow(img_rgb)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(232),plt.imshow(edges,cmap = 'gray')
plt.title('Canny Edge Image(opencv result)'), plt.xticks([]), plt.yticks([])
plt.subplot(233),plt.imshow(imgs_final[0],cmap = 'gray')
plt.title('Canny Edge Image(my implement ,R channel)'), plt.xticks([]), plt.yticks([])
plt.subplot(234),plt.imshow(imgs_final[1],cmap = 'gray')
plt.title('Canny Edge Image(my implement ,G channel)'), plt.xticks([]), plt.yticks([])
plt.subplot(235),plt.imshow(imgs_final[2],cmap = 'gray')
plt.title('Canny Edge Image(my implement ,B channel)'), plt.xticks([]), plt.yticks([])
plt.show()
```
